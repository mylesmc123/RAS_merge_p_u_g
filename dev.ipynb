```python
import pandas as pd
import os
import shutil
```

```python
# Model Name
model_name = 'East_Galveston_Bay'

# Merged Model Direcotry 
merged_model_dir = r"V:\projects\p00659_dec_glo_phase3\01_processing\merge_ras_50_runs\merged\HECRASV6.3.1_BaseModel"
```

```python
# Read csv to dataframe
df = pd.read_csv('GLO RAS Model Runs Distribution to Modelers.csv')
df = df.iloc[1:, 0:6]
df.reset_index(drop=True, inplace=True)
df.dropna(inplace=True)
df
```

```python
for index, row in df.iterrows():
    # print (row['Modeler'])
    # Copy files listed in each Plan_ID, and U Flow File Column to the ./output directory.
    # shutil.copyfile(src, dst)
    plan_file_in = os.path.join(row['Model Location'], f"{model_name}.{row['PlanID']}")
    plan_file_out = os.path.join(merged_model_dir, f"{model_name}.{row['PlanID']}")
    shutil.copy2(plan_file_in, plan_file_out)

    u_flow_file = os.path.join(row['Model Location'], f"{model_name}.{row['U Flow File ID']}")
    u_flow_file_out = os.path.join(merged_model_dir, f"{model_name}.{row['U Flow File ID']}")
    shutil.copy2(u_flow_file, u_flow_file_out)   
    
```
